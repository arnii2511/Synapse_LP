### What is Artificial Neural Network?

Artificial Neural Network ANN is an efficient computing system whose central theme is borrowed from the analogy of biological neural networks. ANNs are also named as “artificial neural systems,” or “parallel distributed processing systems,” or “connectionist systems.” ANN acquires a large collection of units that are interconnected in some pattern to allow communication between the units. These units, also referred to as nodes or neurons, are simple processors which operate in parallel.

Every neuron is connected with other neuron through a connection link. Each connection link is associated with a weight that has information about the input signal. This is the most useful information for neurons to solve a particular problem because the weight usually excites or inhibits the signal that is being communicated. Each neuron has an internal state, which is called an activation signal. Output signals, which are produced after combining the input signals and activation rule, may be sent to other units.

### Components of ANN:
**The Input Layer** — Represents the input variables plus the bias term. Hence if there are n input variables, the size of the input layer is n + 1, where + 1 is the bias term

**The Hidden Layer/ Layers** — These signify neurons where all mathematical calculations are done. Note a given neural network can have more than one neuron in a hidden layer or multiple hidden layers as well

**The Activation Function** — Converts the output of a given layer before passing on the information to consecutive layers. Activation functions are mathematical equations that determine the output of a given neural network. The is a part of each neuron in the hidden layers and determines output relevant for prediction

**The Output Layer** — The final “output prediction” of the network

**Forward Propagation** — Calculating the output of each iteration from the input layer to the output layer

**Backward Propagation** — Calculates revised weights (w1, w2, w3, and b1) after each forward propagation by analyzing the derivative of the cost function used to optimize the model output

**Learning Rate** — Determined the percentage change attributed to each weight and bias term after every backward propagation, i.e. controls the speed at which the model learns information about the data

### Tensor is a multidimensional array of data that represents complex data and is used as a fundamental data structure. Tensors are a generalization of vectors and matrices - Definition of a Tensor agar kisi ko padhna ho toh

### In a neural network, a tensor stores more than just a number or a set of numbers. It's a powerful data structure that encapsulates additional information crucial for the network's learning process. When we perform computations in a neural network, each tensor not only holds its current value but also maintains information about its computational history. This includes storing the gradient of each neuron with respect to the loss function, which is essential for optimizing the network through backpropagation.
### The gradient information allows the network to understand how small changes in each parameter affect the overall loss, guiding the optimization process. Furthermore, tensors in frameworks like TensorFlow also keep track of their "lineage" - the operations and parent tensors from which they were generated. This computational graph structure enables automatic differentiation, where the framework can automatically compute gradients by traversing backwards through the operations that created each tensor.
### This rich structure of tensors, storing values, gradients, and computational history, is what enables efficient and flexible implementation of complex neural network architectures and training algorithms. It's a key reason why specialized libraries like TensorFlow and PyTorch, which are built around the concept of tensors and their operations, are preferred over general-purpose numerical libraries like NumPy for deep learning tasks.



### Upar ke itne bade bhashan se kya samja yaha likho !!!!



lets start with tensor

1. multi-dimensional array (like a matrix, but it can have more than two dimensions).
2. core building block in deep learning because it can hold complex data
3. the complex data include

  1. Values: The actual data (like inputs, weights, or outputs) in multiple dimensions.
  2. Gradients: How 'small changes' in these values affect the overall performance (important for learning).
  3. Computation History: The mathematical operations that were applied to create it (kinda like the backstage of a show but mathematical), helping to compute gradients automatically.

This is also the reason we use tensorflow instead of numpy(this only stores specific data, like very specific compared to a tensor)

In neural networks, we perform operations on these tensors (like matrix multiplication), and these operations help the network learn from data.

So, TensorFlow gets its name because it flows data through these tensors, tracking the operations and gradients automatically, making it ideal for neural network training and optimization.

```python
# If anyone wants to see how a Tensor is defined in Tensorlfow/PyTorch, in the below code you can observe all the functions a tensor needs to perform like calculating and storing the gradient,
# storing its parents and children tensor.

# Yeh code samajho !!!

class Value: # Instead of naming the datatype 'Tensor' it's named 'Value' here
    """ stores a single scalar value and its gradient """

    def __init__(self, data, _children=(), _op=''):
        self.data = data
        self.grad = 0
        # internal variables used for autograd graph construction
        self._backward = lambda: None
        self._prev = set(_children)
        self._op = _op # the op that produced this node, for graphviz / debugging / etc

    def __add__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data + other.data, (self, other), '+')

        def _backward():
            self.grad += out.grad
            other.grad += out.grad
        out._backward = _backward

        return out

    def __mul__(self, other):
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data * other.data, (self, other), '*')

        def _backward():
            self.grad += other.data * out.grad
            other.grad += self.data * out.grad
        out._backward = _backward

        return out

    def __pow__(self, other):
        assert isinstance(other, (int, float)), "only supporting int/float powers for now"
        out = Value(self.data**other, (self,), f'**{other}')

        def _backward():
            self.grad += (other * self.data**(other-1)) * out.grad
        out._backward = _backward

        return out

    def relu(self):
        out = Value(0 if self.data < 0 else self.data, (self,), 'ReLU')

        def _backward():
            self.grad += (out.data > 0) * out.grad
        out._backward = _backward

        return out

    def backward(self):

        # topological order all of the children in the graph
        topo = []
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append(v)
        build_topo(self)

        # go one variable at a time and apply the chain rule to get its gradient
        self.grad = 1
        for v in reversed(topo):
            v._backward()

    def __neg__(self): # -self
        return self * -1

    def __radd__(self, other): # other + self
        return self + other

    def __sub__(self, other): # self - other
        return self + (-other)

    def __rsub__(self, other): # other - self
        return other + (-self)

    def __rmul__(self, other): # other * self
        return self * other

    def __truediv__(self, other): # self / other
        return self * other**-1

    def __rtruediv__(self, other): # other / self
        return other * self**-1

    def __repr__(self):
        return f"Value(data={self.data}, grad={self.grad})"
```

### Create the convolutional base

The 6 lines of code below define the convolutional base using a common pattern: a stack of [Conv2D](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Conv2D) and [MaxPooling2D](https://www.tensorflow.org/api_docs/python/tf/keras/layers/MaxPool2D) layers.

As input, a CNN takes tensors of shape (image_height, image_width, color_channels), ignoring the batch size. If you are new to these dimensions, color_channels refers to (R,G,B). In this example, you will configure your CNN to process inputs of shape (32, 32, 3), which is the format of CIFAR images. You can do this by passing the argument `input_shape` to your first layer.


```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
# Now add another Conv2D followed by a MaxPool and another Conv2D... set the parameters according to what feels right to you and give a justification for the same
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
```

Above, you can see that the output of every Conv2D and MaxPooling2D layer is a 3D tensor of shape (height, width, channels). The width and height dimensions tend to shrink as you go deeper in the network. The number of output channels for each Conv2D layer is controlled by the first argument (e.g., 32 or 64). Typically,  as the width and height shrink, you can afford (computationally) to add more output channels in each Conv2D layer.

### Add Dense layers on top

To complete the model, you will feed the last output tensor from the convolutional base (of shape (4, 4, 64)) into one or more Dense layers to perform classification. Dense layers take vectors as input (which are 1D), while the current output is a 3D tensor. First, you will flatten (or unroll) the 3D output to 1D,  then add one or more Dense layers on top. CIFAR has 10 output classes, so you use a final Dense layer with 10 outputs.

```python
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

## RNN


In a recurrent neural network we store the output activations from one or more of the layers of the network. Often these are hidden later activations. Then, the next time we feed an input example to the network, we include the previously-stored outputs as additional inputs. You can think of the additional inputs as being concatenated to the end of the “normal” inputs to the previous layer.

A simple explanation of RNN : https://www.youtube.com/watch?v=UNmqTiOnRfg&t=3s

RNN Overview

Recurrent Neural Networks (RNNs):
1. Deep learning models for sequential data (e.g., text, time-series, speech)
2. Retain memory of previous inputs to improve predictions
3. Ideal for language translation, speech recognition, and trend forecasting.

Working
1. RNNs process inputs one step at a time, retaining context from earlier steps.
2. Memory + Predictions: A hidden layer stores previous inputs, improving predictions in sequences (e.g., understanding “Apple is” to predict “red”).
3. Self-Looping (Recurrence): The network feeds previous outputs into the next step, enabling it to learn relationships between sequence elements.

Limitations
1. Exploding/Vanishing Gradient: Gradients can either grow too large (explode)(leading to instability) or shrink (vanish)(leading to poor learning), especially in long sequences.
2. Slow Training: RNNs process data sequentially, making training time-consuming, particularly with large datasets.
3. Limited Long-Term Memory: RNNs might not retain distant information very well, affecting their effectiveness in complex tasks requiring long-term dependencies.

Comparing with-
1. Feed-Forward Networks (FNNs): RNNs retain input memory, FNNs don't. FNNs are suited for independent input tasks like image classification.
2. Convolutional Neural Networks (CNNs): RNNs handle sequential data(text,etc); CNNs excel at spatial data (images).

How can i overcome the limitations of RNNs
1. Using LSTMs: Designed to overcome gradient issues and better handle long-term dependencies.
2. Transformers: A newer, faster model that processes sequences in parallel, replacing RNNs for many tasks.

What is LSTM?

Long Short-Term Memory (LSTM) is a type of Recurrent Neural Network (RNN) that helps in remembering information over long periods. It was developed by Hochreiter and Schmidhuber to handle sequences of data more effectively and avoid problems like forgetting important information.

LSTM Architecture

LSTMs have a memory cell and three main gates that control the flow of information:

1. Input Gate: Decides how much new information to store in the memory cell.
2. Forget Gate: Determines which information to discard from the memory cell.
3. Output Gate: Decides what information to pass on to the next layer.

Types:

1. Bidirectional LSTMs (BiLSTM)

BiLSTMs process data in both directions (forward and backward), improving their ability to learn from context and enhancing performance in tasks like machine translation and speech recognition.

2. Stacked LSTM

LSTMs can be stacked in multiple layers, allowing them to learn more complex patterns in data by capturing different levels of detail.

Key Applications

LSTMs are widely used in:

1. Time Series Forecasting: Predicting future values like stock prices and weather.
2. Language Translation: Translating text and generating coherent sentences.
3. Speech Recognition: Converting spoken words to text and recognizing commands.
4. Text Summarization: Condensing long texts into key points.
5. Anomaly Detection: Identifying unusual patterns in data, like fraud.
6. Recommender Systems: Personalizing content suggestions based on past user behavior.
7. Video Analysis: Detecting objects and recognizing actions in videos.

1. Gated Recurrent Unit (GRU)
- **Definition**: A simplified version of LSTMs that combines the forget and input gates into a single update gate. It has fewer parameters than LSTMs but is able to learn long-term dependencies.
- **Advantages**: Faster to train than LSTMs and often performs similarly well in practice.

2. Attention Mechanisms
- **Description**: While not an RNN per se, attention mechanisms can be integrated with RNNs to help the model focus on relevant parts of the input sequence. This has become a critical component in many advanced models.
- **Advantages**: Allows the model to selectively focus on certain inputs when making predictions, improving performance in tasks like translation and summarization.

3. Bidirectional RNN (BiRNN)
- **Description**: This is an RNN that processes the input sequence both forward and backward directions. It takes into account information from both previous and subsequent contexts.
- **Advantages**: It performs better on those tasks in which the context from both sides is required or important. Text classification and machine translation fall into that category.
