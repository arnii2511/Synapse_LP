````python
# Write a recursive function to check if a given string is a palindrome or not.

def isPalindrome(stt,i):
  if (i>len(stt)//2):
    return True
  res=False
  if ((stt[i] is stt[len(stt)-i-1]) and isPalindrome(stt,i+1)):
    res=True
  return res

stt=input("Enter a string: ")
if isPalindrome(stt,0)==True:
  print("Yes, it is a palindrome")

else:
  print("No, it is not a palindrome")
````

````python
# Write a class in Python for a point (1D)
# Use functions like __str__ for it. Find out what they do ;)
class Point:
  def __init__(self,a,b):
    self.a=a
    self.b=b

  def __str__(self):
    return (f"("+str(self.a)+","+str(self.b)+")")

p1=Point(1,2)
p2=Point(9,10)
print(p1)
print(p2)

# Extend it for a Line. (2D)
class Line:
  def __init__(self,p1,p2):
    self.p1=p1
    self.p2=p2

  def __str__(self):
    return (f"("+str(self.p1)+","+str(self.p2)+")")

l1=Line(p1,p2)
print(l1)
````

###Armstrong number

Armstrong number is a number that is equal to the sum of the nth power of its digits. Here, n is the number of digits of the Armstrong number.

For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

Say, 153.
1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153!
Armstrong.

````python
# Follow good practices and write code that accepts start and limit integers,
# and returns a list of Armstrong numbers between them (both inclusive).
import math

def armstrong(start,end):

  for i in range (start,end+1):
    x=i
    n=0
    while (x>0):
      x=x//10
      n=n+1
    ps = 0
    x=i
    while (x!=0):
      d=x%10
      ps = ps + int(math.pow(d,n))
      x=x//10
    if ps == i:
      print(str(i)+" ")



start=int(input("Enter start: "))
end=int(input("Enter end: "))
armstrong(start,end)




````

###Collatz sequence

Given a number, if it is even, the next number in collatz sequence is n/2. If it is odd, the next number is 3*n + 1.

If you do this enough times, you will DEFINITELY end with a sequence of 4, 2, 1.

(Search up how later, confusing but interesting)

Return the collatz sequence of the input number, ending with the 4, 2, 1.

````python
# Taking one input number, return its Collatz sequence in a list.
# Ending with the 4, 2, 1

n=int(input("Enter a number"))
def collatz(n):
  lst=[]
  lst.append(n)
  while n!=1:
    if n%2==0:
      n=n//2
      lst.append(n)
    else:
      n=3*n+1
      lst.append(n)
  return lst

print(collatz(n))
````

A box plot is also known as a box and whisker plot. It is used for graphical representation of a dataset. It uses five key statistics points to show the data, hence giving a pretty good analysis of the graph and data.

The 5 key statistics points of Box Plot are
1. Box:
a. Median = The line inside the box is the median (50th percentile) and divides the data into 2 equal halves.
b. Interquartile Range- This range is from 25 percentile to 75 percentile. Basically the range in which the central 50% of data falls.

2. Whiskers:
a. Upper - The long vertical line above the box that extends from 75 percentile mark to the 100 percentile mark (maximum value)
b. Lower - The long vertical line below the box that extends down from 25 percentile mark to the 0 percentile mark (minimum value)

3. Outliers: These can be considered kind of an anomaly or exception in your datasetas they differ significantly from the other values in the data. they might be unusually high or low. They are marked outside the whisker mark and plotted in points.

Interpreting the graph
1. Central Tendency= This is the Median Line giving us the central value.

2. Spread = Length of the box shows the spread of the 50% central data.

3. Skewness = Shows assymetry in the distribution if any. happens when the median is situated more closer to one end of the box.

4. Outliers = tells us how many unusual or exceptional values are there in the dataset.
